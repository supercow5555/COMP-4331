```python
import numpy as np 
import pandas as pd
from sklearn import tree
from sklearn.metrics import accuracy_score
from sklearn import preprocessing
```

## Data preprocessing

```python
#Loading data 
data_file = "data.csv"
data_df = pd.read_csv(data_file, sep=",")
```

```python
cols = ["age", "income", "student", "credit_rating", "class_buys_computer"]
```

```python
#Doing data preprocessing 
#Creating dummy varialbes 
from sklearn.preprocessing import OrdinalEncoder

enc = OrdinalEncoder(dtype = int)
enc.fit(data_df[cols])
attributes = enc.transform(data_df[cols])
```

```python
#Getting x (attributes) and y(classes)
x_data = attributes[:, 0:4]
y_data = attributes[:, 4:5]
```

## Question 1a

```python
#Training decision tree 
x_train = x_data[4:14]
y_train = y_data[4:14]
clf = tree.DecisionTreeClassifier(criterion="entropy", min_samples_split=2, random_state=1)

clf=clf.fit(x_train, y_train)
```

```python
#Plotting decision tree 
import matplotlib.pyplot as plt 
plt.figure(figsize=(20, 10))
tree.plot_tree(clf, feature_names=cols)
plt.show()
```

```python
#Finding the training and testing accurac 
train_score = accuracy_score(y_true = y_train, y_pred = clf.predict(x_train))
print("The training accuracy is: ", train_score)

x_test = x_data[0:4]
y_test = y_data[0:4]
test_score = accuracy_score(y_true = y_test, y_pred = clf.predict(x_test))
print("The testing accuracy is: ", test_score)
```

```python
y_train_pred = clf_NB.predict(x_train_NB)
y_test_pred = clf_NB.predict(x_test_NB)

train_acc = (sum(y_train_pred == y_train_NB) / len(y_train_NB))
test_acc = (sum(y_test_pred == y_test_NB)/ len(y_test_NB))

print("The training accuracy is: ", train_acc)
print("The testing accuracy is: ", test_acc)
```
