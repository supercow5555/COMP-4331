```python
import re
def extract_credit(text):
    text = text.lower() # lower text

    # find mult
    if "iou" in text:
        mult = -1
    elif "uome" in text:
        mult = 1
    else:
        mult = 0
    
    # find val
    val = re.search(r"[0-9]+(\.[0-9]+)?",text)
    try:
        val = float(val.group(0))
    except Exception as e:
        print(f"{text = }, {val = }")
        print(e)
        val = 0

    return val*mult
```

```python
raw_df["credit"] = raw_df['Message'].apply(extract_credit)
```
