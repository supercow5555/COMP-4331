```python
import re
def extract_credit(text):
    text = text.lower() # lower text

    # find mult
    if "iou" in text:
        mult = -1
    elif "uome" in text:
        mult = 1
    else:
        mult = 0
    
    # find val
    val = re.search(r"[0-9]+(\.[0-9]+)?",text)
    try:
        val = float(val.group(0))
    except Exception as e:
        print(f"{text = }, {val = }")
        print(e)
        val = 0

    return val*mult
```

```python
print(extract_credit("UoMe 78.63374"))
```

```python
raw_df["credit"] = raw_df['Message'].apply(extract_credit)
```

```python
import numpy as np
raw_df["kk_gain"] = np.where(raw_df["Name"] == "Saumya", raw_df["credit"]*-1, raw_df["credit"])
```
